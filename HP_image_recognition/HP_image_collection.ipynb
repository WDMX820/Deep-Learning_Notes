```python
# Harry Potter image collection

# 使用前请对应浏览器版本（edge://version/）下载对应的 WebDriver
# 安装 bing-image-downloader
!pip install bing-image-downloader

# 导入库
from bing_image_downloader import downloader
import os
from concurrent.futures import ThreadPoolExecutor

# 获取桌面路径 (以 Windows 为例)
desktop_dir = os.path.join(os.path.expanduser("~"), "Desktop")

# 定义角色列表
characters = [
    "Harry Potter", "Hermione Granger", "Ron Weasley",
    "Albus Dumbledore", "Voldemort", "Severus Snape",
    "Rubeus Hagrid"
]

# 定义单角色下载函数
def download_images_for_character(character, num_images=100):
    save_dir = os.path.join(desktop_dir, "harry_potter_characters_1")  # 保存到桌面文件夹
    
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)
    
    character_dir = os.path.join(save_dir, character.replace(" ", "_"))
    if not os.path.exists(character_dir):
        os.makedirs(character_dir)
    
    # 下载图片
    print(f"Downloading images for {character}...")
    downloader.download(character, limit=num_images, output_dir=save_dir, 
                        adult_filter_off=True, force_replace=False, timeout=15)  # 设置较小的超时时间
    print(f"Downloaded {num_images} images for {character}.")

# 使用多线程并发下载
def download_images_for_characters_concurrently(characters, num_images=100):
    with ThreadPoolExecutor(max_workers=3) as executor:  # 设置同时下载的线程数为 3
        futures = [executor.submit(download_images_for_character, character, num_images) for character in characters]
        for future in futures:
            future.result()  # 等待所有任务完成

# 下载哈利波特角色的照片，每个至少100张
download_images_for_characters_concurrently(characters, num_images=100)
```
